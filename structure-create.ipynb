```python
import os

structure = {
    "01_Introduction": [
        "01_Qu_est_ce_que_l_automatisation.md",
        "02_Pourquoi_automatiser.md",
        "03_Domaines_d_application.md",
        "04_Pre_requis.md"
    ],
    "02_Fondamentaux": [
        "01_Concepts_cles.md",
        "02_Types_d_automatisation.md",
        "03_Identifier_les_taches.md",
        "04_Securite_et_ethique.md"
    ],
    "03_Outils_et_Technos": [
        "01_Langages_de_scripting.md",
        "02_Outils_workflow.md",
        "03_Planification.md",
        "04_Orchestration.md",
        "05_Outils_tests.md"
    ],
    "04_Mise_en_pratique": [
        "01_Fichiers_et_dossiers.md",
        "02_Interactions_web.md",
        "03_Ligne_de_commande.md",
        "04_Communications.md",
        "05_Bureautique.md"
    ],
    "05_Automatisation_avancee": [
        "01_IA_et_Ollama.md",
        "02_Infrastructure_as_code.md",
        "03_Tests_logiciels.md",
        "04_Gestion_des_donnees.md"
    ],
    "06_Production_et_maintenance": [
        "01_Execution_et_deploiement.md",
        "02_Surveillance_et_logs.md",
        "03_Maintenance_et_evolution.md"
    ],
    "07_Conclusion": [
        "01_Prochaines_etapes.md",
        "02_Ressources_et_communautes.md"
    ],
    "08_Annexes": [
        "01_Scripts_reutilisables.md",
        "02_Outils_recommandes.md",
        "03_Glossaire.md"
    ]
}

# Création de la structure
for folder, files in structure.items():
    os.makedirs(folder, exist_ok=True)
    for file in files:
        filepath = os.path.join(folder, file)
        with open(filepath, "w", encoding="utf-8") as f:
            f.write(f"# {file.replace('_', ' ').replace('.md', '')}\n")

# Création du README principal
with open("README.md", "w", encoding="utf-8") as readme:
    readme.write("# Projet Tutoriels : Automatisation des tâches\n")
    readme.write("\nCe dépôt contient des tutoriels organisés autour de l'automatisation informatique.\n\n")
    readme.write("## Sommaire\n")
    for folder in sorted(structure):
        title = folder[3:].replace('_', ' ')
        readme.write(f"- [{title}](./{folder})\n")
```
